```python
import requests
from bs4 import BeautifulSoup

def soup_url(url): 
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib')
    return soup

result_url = soup_url("https://epyelectronica.com/")
```

```python
extract_urls = str(result_url.find_all('ul', class_='sub-menu'))
soup_urls = BeautifulSoup(extract_urls, 'lxml')
list_products = [i['href'] for i in soup_urls.find_all('a', href=True)]
for i,url in enumerate(list_products[:54]):
    print(str(i) +  " Processing: " + url)
    result_url = soup_url(url)
    titles = result_url.find_all('a', class_='woocommerce-LoopProduct-link-title woocommerce-loop-product__title_ink')
    prices = result_url.find_all('span', class_='woocommerce-Price-amount amount')
    for product, price in zip(titles,prices):
        print(product.text)
        print(price.text)
```
